- 해석 포인트: long-read 결과는 isoform 구조 해석에 유리하지만, 정량 정확도는 read depth와 preprocessing 조건에 영향을 받습니다.

### __2. minimap2 align__

#### __Example 1: PacBio long-reads for NSCLC cellline__

```python
src_dir = 'RNAseq_pacbio_chr7'
flst = os.listdir(src_dir)
flst.sort()
flst
```

```python
df_fq_pair_list = utl.get_fq_pair_list( src_dir, ext = 'fq.gz',
                                        r1_key = None, r2_key = None )
df_fq_pair_list
```

```python
df_fq_pair_list.iloc[2:]
```

```python
## Only for test DEG analysis
df_fq_pair_list['RG'] = 'N'
df_fq_pair_list.loc[[2,3],'RG'] = 'T'

df_fq_pair_list['SM'] = 'N'
df_fq_pair_list.loc[[2,3],'SM'] = 'T'

df_fq_pair_list
```

```python
path_to_idx = "index_mm2_hg38_sel"
out_dir = "RNAseq_pacbio_chr7_bam"
if not os.path.isdir(out_dir):
    os.mkdir(out_dir)

df_fq_pair_list['bam'] = None
for j, row in df_fq_pair_list.iterrows():

    fq_l = row.fq1
    out_file = '_'.join( fq_l.split('/')[-1].split('_')[:-1] )

    if (row.RG is not None) & (row.SM is not None):
        other_option = f'-R \'@RG\\tID:{row.RG}\\tSM:{row.SM}\\tPL:PacBio\''
    else:
        other_option = ''

    print(j, out_file)

    file_out, cmd_lst = aln.align_n_bam_sort_minimap2(
                              fq_left = fq_l,
                              path_to_idx = path_to_idx,
                              out_dir = out_dir,
                              fq_right = None,
                              out_filename = None,
                              p = 4,
                              x_opt = 'splice:hq',
                              other_opt = other_option,
                              verbose = True )

    df_fq_pair_list.loc[j, 'bam'] = file_out
    print('Output file saved to %s' % file_out)
```

## __3. Gene/Transcript expression quantification__

<!-- MLBI-EDU-NOTE: practice -->
이 단계에서는 assembled transcript expression을 이용해 조건 간 차이를 확인합니다. 예제는 교육용 subset이므로 p-value 자체보다 분석 객체의 구조와 비교 방향을 이해하는 데 초점을 둡니다.


#### __A. DEG__

```python
t_or_g = 't'
method_selection = 'pb_strg_strg' ## 'salmon' or 'rsem'

df_g_exp = pd.read_csv( f'exp_{t_or_g}_{method_selection}.csv', index_col = 0 )
df_tr_exp = pd.read_csv( f'exp_{t_or_g}_{method_selection}.csv', index_col = 0 )
```

```python
df_expr = np.log2( 1 + df_g_exp )
groups = [c.split('-')[0] for c in df_expr.columns]
np.array(groups)
```

```python
groups = ['T', 'T', 'N', 'N']
```

```python
dct_deg, dct_deg_all = deg.simple_DEG( df_expr, groups,
                                       ref_condition = 'N', ## None for One-vs-the-Rest
                                       pval_cutoff = 0.05,
                                       logFC_cutoff = 0.5,
                                       test_method = "ols",
                                       corr_method = "fdr_bh")

## dct_deg contains only significant DEGs only
## dct_deg_all contains in-significant genes as well

dct_deg.keys()
```

```python
item = 'T_vs_N'
dct_deg[item].to_csv(f'deg_{t_or_g}_{method_selection}.csv')
```

```python
dct_deg[item].head()
```

```python
score_ref = 'logFC' # 'pval_adj' or 'logFC'
deg.plot_deg( dct_deg, reference = score_ref, n_genes_to_show = 30, pval_cutoff = 0.05,
              figsize = (6,4), dpi = 100, text_fs = 10, title_fs = 12, label_fs = 11,
              tick_fs = 10, ncols = 2, wspace = 0.15, hspace = 0.2,
              deg_stat_dct = None, show_log_pv = True )
```

```python
deg.plot_volcano(
    dct_deg_all[item],
    logFC_col = "logFC",
    pval_col = "pval_adj",
    gene_col = None,
    pval_cutoff = 0.05,
    logFC_cutoff = 1.0,
    top_n_labels = 15,
    figsize=(6, 6),
    # xlim = [-5, 5],
    # ylim = [0, 3],
    title = "Volcano plot",
)
```

## __5. Quantifying alternative splicing using RMATS__

<!-- MLBI-EDU-NOTE: practice -->
이 섹션은 short-read RNA-seq BAM을 이용한 rMATS 예제를 참고용으로 포함합니다. Long-read 기반 isoform 분석과 short-read 기반 event 분석이 어떤 관점 차이를 갖는지 비교해 보세요.


```python
!rmats.py --version
```

```python
src_dir = 'RNAseq_CRC_chr17_bam'
flst = os.listdir(src_dir)
flst.sort()
flst = [s for s in flst if s.split('.')[-1] == 'bam']
flst
```

```python
df_bam_pair_lst = utl.get_bam_pair_list( src_dir, ext = 'bam' )
df_bam_pair_lst
```

#### __A. Prepare input BAM list__

```python
## prepare bam list file
group_lst = ['bcatKO', 'vector', 'parent']
f_lst_dct = {}
for g in group_lst:
    s = ','.join( df_bam_pair_lst.loc[ df_bam_pair_lst['tumor_RG'] == g, 'tumor_bam'] )
    fn = f'b_{g}.txt'
    f_lst_dct[g] = fn
    with open(fn, 'w') as f:
        f.writelines(s)
f_lst_dct
```

#### __B. Run RMATs__

```python
t1, t2 = 'bcatKO', 'vector'

gtf_file = 'RefGenome_hg38_sel/Homo_sapiens_assembly38_chr7_12_17.gtf'

out_dir = src_dir + '_rmats'
if not os.path.isdir(out_dir):
    os.mkdir(out_dir)
```

```python
b1 = f_lst_dct[t1]
b2 = f_lst_dct[t2]

tasm.rMATs_run( b1, b2, gtf_file, rd_type = 'paired',
                rd_len = 100, n_thred = 4, out_dir = out_dir,
                lib_type = 'fr-firststrand', verbose = True )
```

#### __C. Get results in data frames__

```python
df_summary, df_rmats, df_results = tasm.rMATs_summarize_result( out_dir, pval_cutoff = 0.05 )
```
